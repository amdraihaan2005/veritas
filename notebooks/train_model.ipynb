```python
import pandas as pd
import numpy as np

from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler

from sklearn.metrics import (
    confusion_matrix,
    classification_report,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score
)

import joblib
```

```python
df = pd.read_csv("../data/creditcard.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
X = df.drop("Class", axis=1)
y = df["Class"]

print(X.shape)
print(y.shape)
```

```python
X_train = X[y == 0]

print(X_train.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_train_scaled.shape
```

```python
model = IsolationForest(
    n_estimators=100,
    contamination=0.005,
    random_state=42,
    n_jobs=-1
)
```

```python
model
```

```python
model.fit(X_train_scaled)
```

```python
X_scaled = scaler.transform(X)

X_scaled.shape
```

```python
predictions = model.predict(X_scaled)
```

```python
np.unique(predictions, return_counts=True)
```

```python
y_pred = np.where(predictions == -1, 1, 0)

np.unique(y_pred, return_counts=True)
```

```python
cm = confusion_matrix(y, y_pred)

cm
```

```python
precision = precision_score(y, y_pred)
recall = recall_score(y, y_pred)
f1 = f1_score(y, y_pred)

print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```

```python
from sklearn.metrics import classification_report

print(classification_report(y, y_pred))
```

```python
anomaly_scores = model.decision_function(X_scaled)
```

```python
anomaly_scores[:10]
```

```python
risk_scores = ((anomaly_scores.max() - anomaly_scores) / (anomaly_scores.max() - anomaly_scores.min())) * 100
```

```python
risk_scores[:10]
```

```python
pd.Series(risk_scores).describe()
```

```python
results_df = df.copy()

results_df["Risk_Score"] = risk_scores

results_df.sort_values(by="Risk_Score", ascending=False).head(10)
```

```python
top_100 = results_df.sort_values(
    by="Risk_Score",
    ascending=False
).head(100)

top_100["Class"].value_counts()
```
